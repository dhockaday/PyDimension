# Test the reshape of Flow3D data

```python
import pandas as pd
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
```

```python
# printing vortz       t=0.07200000  ix=2 to  501   jy=121 to  121  kz=2 to
file_path = '/home/xie/projects/DimensionNet/dataset/data/13-Re99-cylinder/Line_Horizontal_Reshape.txt'
data_all = []
with open(file_path, 'r') as f:
    lines = f.read().splitlines() 
    for line in lines:
        str_split = line.split(' ')
        str_clean = [float(each) for each in str_split if each != '']
        data_all.append(str_clean)

df_line = pd.DataFrame(data_all, columns=['x', 'y', 'z', 'u'])
```

```python
Nx, Nz, Nt = 500, 222, 201
w_3D = np.load('../dataset/data/13-Re99-cylinder/w_3D.npy')

fig = plt.figure()
plt.plot(range(Nx), w_3D[0, 121, :], label='rearanged')
plt.plot(range(Nx), df_line['u'], label='true')
plt.legend(fontsize=14)
```
